```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1  = pd.read_csv('./Complementary/air_quality_data.csv')
df1.head()
```

```python
df1.shape
```

```python
# Convertir la columna 'dt' a formato de fecha
df1['dt'] = pd.to_datetime(df1['dt'])
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='dt', y='co', data=df1, palette='colorblind')
plt.title('Grfico de barras, concentracin de CO por da')
plt.xlabel('Da')
plt.ylabel('Concentracin de CO2')
plt.show()
```

- Aqui se puede ver que el rango de valores para el co es mucho mayor que para los demas se hacen dos graficos, uno solo del co y los otros en otro grafico

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el grfico de lneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df1, x='dt', y='co', marker='o', label='CO')


plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.title('Variacin de CO')
plt.legend()
plt.xticks(rotation=45)

# Mostrar el grfico
plt.show()
```

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el grfico de lneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df1, x='dt', y='no', marker='o', label='NO')
sns.lineplot(data=df1, x='dt', y='no2', marker='o', label='NO2')
sns.lineplot(data=df1, x='dt', y='o3', marker='o', label='O3')
sns.lineplot(data=df1, x='dt', y='so2', marker='o', label='SO2')
sns.lineplot(data=df1, x='dt', y='pm2_5', marker='o', label='PNM2_5')
sns.lineplot(data=df1, x='dt', y='pm10', marker='o', label='PM10')
sns.lineplot(data=df1, x='dt', y='nh3', marker='o', label='NH3')

plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.title('Variacin de Gases con el tiempo')
plt.legend()
plt.xticks(rotation=45)

# Mostrar el grfico
plt.show()
```

Se observa con en 2021 hubo menos contaminacion por la pandemia, pero me parece que las fechas que tenemos no son sufientes, porque vienen solo desde el 2021 hasta el 2022

La calidad de la data esta correcta y no hace falta ningn valor en las columnas, recordar agregar estos valores al diccionario.

```python
# Calcular la matriz de correlacin
correlation_matrix = df1.corr()

# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el pair plot
sns.pairplot(df1, diag_kind="kde")

# Mostrar el pair plot
plt.show()

# Crear la matriz de correlacin
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)

plt.title("Matriz de Correlacin")
plt.show()
```
