Teniendo en cuenta que el alcance del proyecto incluye aspectos relacionados con contaminaciÃ³n ambiental y que se podrÃ­an analizar cuestiones relacionadas con costos, es que se decide trabajar con un dataframe reducido:

Dado que saber el modelo del vehÃ­culo es crucial para el anÃ¡lisis, las filas que no tienen este datos, no serÃ¡n tenidas en cuenta.

De la misma forma, tampoco serÃ¡n tenidas en cuenta las filas que no tengan datos de co2TailpipeGpm (emisiÃ³n de CO2 medida en el caÃ±o de escape en gramos por milla)

```python
fuel_economy_reducido.dropna(subset=['Model', 'co2TailpipeGpm'], inplace=True)
fuel_economy_reducido.isnull().sum()
```

RevisiÃ³n de valores en columnas de texto

1. Tipo de vehÃ­culo

```python
sorted(fuel_economy_reducido['VClass'].astype(str).unique())
```

Para simplificar:

Minivan - 2WD y Minivan - 4WD => Minivan

Small Pickup Trucks 2WD y Small Pickup Trucks 4WD => Small Pickup Trucks

Standard Pickup Trucks 2WD, Standard Pickup Trucks 4WD, Standard Pickup Trucks/2wd => Standard Pickup Trucks

Small Sport Utility Vehicle 2WD y Small Sport Utility Vehicle 4WD  => Small Sport Utility Vehicle

Standard Sport Utility Vehicle 2WD y Standard Sport Utility Vehicle 2WD => Standard Sport Utility Vehicle

Sport Utility Vehicle - 2WD y Sport Utility Vehicle - 4WD => Sport Utility Vehicle

Special Purpose Vehicle 2WD, Special Purpose Vehicles/2wd, Special Purpose Vehicle 4WD y Special Purpose Vehicles/4wd =>  Special Purpose Vehicle

```python
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Minivan' if x== 'Minivan - 2WD' or  x == 'Minivan - 4WD' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Small Pickup Trucks' if x== 'Small Pickup Trucks 2WD' or  x == 'Small Pickup Trucks 2WD' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Standard Pickup Trucks' if x== 'Standard Pickup Trucks 2WD' or  x == 'Standard Pickup Trucks 4WD' or x == 'Standard Pickup Trucks/2wd' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Small Sport Utility Vehicle' if x== 'Small Sport Utility Vehicle 2WD' or  x == 'Small Sport Utility Vehicle 4WD' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Standard Sport Utility Vehicle' if x== 'Standard Sport Utility Vehicle 2WD' or  x == 'Standard Sport Utility Vehicle 4WD' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Sport Utility Vehicle' if x== 'Sport Utility Vehicle - 2WD' or  x == 'Sport Utility Vehicle - 4WD' else x )
fuel_economy_reducido['VClass'] = fuel_economy_reducido['VClass'].apply(lambda x : 'Special Purpose Vehicle' if x== 'Special Purpose Vehicle 2WD' or  x == 'Special Purpose Vehicle 4WD' or x== 'Special Purpose Vehicles/2wd' or x == 'Special Purpose Vehicles/4wd' else x )

```

```python
fuel_economy_reducido['drive'] = fuel_economy_reducido['drive'].apply(lambda x : '4-Wheel Drive' if x== '4-Wheel or All-Wheel Drive' or  x == 'All-Wheel Drive' else x )
fuel_economy_reducido['drive'] = fuel_economy_reducido['drive'].apply(lambda x : '2-Wheel Drive' if x== 'Rear-Wheel Drive' or  x == 'Front-Wheel Drive' else x )
```
